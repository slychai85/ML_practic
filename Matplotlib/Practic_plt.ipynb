```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Нарисовать функции косинуса и синуса на одном графике.

```python
# создание данных
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

`X` теперь является массивом с 256 значениями в диапазоне от `-π` до `+π` (включительно). `C` - это косинус (256 значений), а `S` - это синус (256 значений).

```python
# дефолтные настройки
plt.plot(X, C)
plt.plot(X, S)
```

```python
# Создание фигуры размером 10x6 дюймов, 60 точек на дюйм
plt.figure(figsize=(10, 6), dpi=60)

# Создать новый участок из сетки 1x1
plt.subplot(1, 1, 1)

# Построить косинус с синей непрерывной линией шириной 2 (в пикселях)
plt.plot(X, C, color='blue', linewidth=2.0, linestyle='-')
plt.plot(X, S, color='green', linewidth=3.5, linestyle=':')

# Установить ограничения по оси Х
# plt.xlim(-3.5, 3.5)
plt.xlim(X.min() * 1.1, X.max() * 1.1)

# Установить риски на оси X. Можно создать название и подать массив
# plt.xticks(np.linspace(-3, 3, 7, endpoint=True))
# plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
# Замена данных насечек на более понятные и читаемые
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Установить ограничения по оси Y
# plt.ylim(-1.05, 1.05)
plt.ylim(C.min() * 1.1, C.max() * 1.1)

# Установить риски на оси Y
# plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
# plt.yticks([-1, 0, +1])
# Замена данных насечек на более понятные и читаемые
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

plt.show()
```

Рамки (spines) - это линии, соединяющие метки осей и отмечающие границы области данных. Их можно размещать в произвольных положениях и до сих пор они находились на границе оси. Мы изменим это, так как хотим, чтобы они были посередине. Поскольку их четыре (верхний / нижний / левый / правый), мы отбросим верхний и правый, установив для них цвет «none», и переместим нижний и левый в координату 0 в координатах пространства данных.

```python
# Создание фигуры размером 8x5 дюймов, 80 точек на дюйм
plt.figure(figsize=(8, 5), dpi=80)
plt.subplot(111)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-",  label="sine")

# gca означает «получить текущую ось»
# этот блок отвечает за то, чтобы убрать рамки побокам и сделать из
# в центре как оси графики
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

# Установить ограничения по оси Х
plt.xlim(X.min() * 1.1, X.max() * 1.1)
# Установить риски на оси X
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Установить ограничения по оси Y
plt.ylim(C.min() * 1.1, C.max() * 1.1)
# Установить риски на оси Y
plt.yticks([-1, 1], [r'$-1$', r'$+1$'])

# Добавить легенду в верхнем левом углу.
plt.legend(loc='upper left')

# Анотация некоторых точек
t = 2*np.pi/3
plt.plot([t, t], [0, np.cos(t)],
        color='blue', linewidth=1.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
            xy=(t, np.sin(t)), xycoords='data',
            xytext=(10, 30), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t, t], [0, np.sin(t)],
        color='red', linewidth=1.5, linestyle="--")
plt.scatter([t, ], [np.sin(t), ], 50, color ='red')
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy=(t, np.cos(t)),
            xycoords='data', xytext=(-90, -50),
            textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

# Метки тиков теперь плохо видны из-за синих и красных линий. 
# Мы можем сделать их больше и настроить их свойства так, 
# чтобы они отображались на полупрозрачном белом фоне. 
# Это позволит нам увидеть как данные, так и ярлыки.
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))

# При желание можно сохранить этот рисунок c определенным разрешением
# plt.savefig("exercise.png", dpi=80)

plt.show()
```
